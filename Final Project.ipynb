```python
def remove_punct(text):
    text_nopunct = "".join([char for char in text if char not in string.punctuation])
    return text_nopunct

posts['body_text_clean'] = posts['sharemediatext'].apply(lambda x: remove_punct(x))

posts.head()
```

```python
import re

def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens

posts['body_text_tokenized'] = posts['body_text_clean'].apply(lambda x: tokenize(x.lower()))

posts.head()
```

```python
import nltk

stopword = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(tokenized_list):
    text = [word for word in tokenized_list if word not in stopword]
    return text

posts['body_text_nostop'] = posts['body_text_tokenized'].apply(lambda x: remove_stopwords(x))

posts.head()
```

```python
ps = nltk.PorterStemmer()
```

```python
def stemming(tokenized_text):
    text = [ps.stem(word) for word in tokenized_text]
    return text

posts['body_text_stemmed'] = posts['body_text_nostop'].apply(lambda x: stemming(x))

posts.head()
```
